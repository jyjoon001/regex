```
import glob

filedir = glob.glob('/content/drive/MyDrive/Exobrain/corpus4/*')
print(filedir)
```

```
sentcat = []
for i in range(0, len(filedir)-1):
    data = open(filedir[i], 'r', encoding='utf-8')
    sent = data.readlines()
    sentcat.append(sent)
```

```
import re
def clean_text(text):
    for sentemp in text:
        pattern0='"'
        text = re.sub(pattern=pattern0, repl='', string=text)
        pattern0='\n'
        text = re.sub(pattern=pattern0, repl='', string=text)
        text = text.replace("\\", '"')
        text = text.replace(".,", '.')       
        text = text.replace("?,", '?')            
    return text    
```

```
cleaned = []
for i in range(0, len(sentcat)-1): 
    for j in range(0, len(sentcat[i])-1):
      if xword in sentcat[i][j]:
        pass
      elif xword2 in sentcat[i][j]:
        pass
      else:
        cleaned.append(clean_text(sentcat[i][j]))
```

```
cleaned = [ele for ele in cleaned if ele != ''] 
res = [] 
[res.append(x) for x in cleaned if x not in res] 
```

```
# res.sort()
print(len(res))
for i in range(0, 10):
    print(str(res[i]))
```

```
import pandas as pd
corpus = [res]
df = pd.DataFrame (corpus,index=[''])
print(df.T)
```
